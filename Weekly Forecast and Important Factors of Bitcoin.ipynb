# การทำนายรายสัปดาห์และปัจจัยที่สำคัญของ Bitcoin  
  
* Bitcoin เป็นสินทรัพย์ทางดิจิทัลที่ได้รับความนิยมและถูกพูดถึงมากในช่วงทศวรรษนี้ โดยมีจุดประสงค์คือ การสร้างสกุลเงินที่เป็นอิสระจากรัฐบาลและธนาคารสามารถส่งหากันผ่านระบบอินเทอร์เน็ตโดยไม่จำกัดขอบเขตและมีค่าธรรมเนียมที่ถูกมาก ต้นทุนที่ต่ำ และ ความพิเศษของ Bitcoin ที่เป็นตัวช่วยให้มันเป็นที่นิยมคือมันถูกควบคุมแบบกระจาย(decentralize) กล่าวคือ ไม่มีสถาบันการเงินไหนสามารถควบคุมบิทคอยได้ ซึ่งนั่นทำให้ผู้คนที่เลือกใช้และได้รับการยอมรับมากขึ้น  
  
  
* โปรเจคนี้ต้องการจะศึกษาปัจจัยที่สำคัญที่มีผลต่อราคาBitcoin และ ทำนายBitcoinเป็นรายสัปดาห์ เพราะ ทางผู้เขียนเล็งเห็นว่า ปัจจุบันมีผู้ให้วามสนใจกับcrytocurrencyเป็นจำนวนมากโดยเฉพาะbitcoin และมีผู้ที่เสียเงินเป็นจำนวนมากในการลงทุนเช่นกันเดียวกัน เพราะ แต่ละคนก็มีทักษะและความสามารถในการวิเคราะห์แตกต่างกัน ผู้เขียนจึงอยากจะลองศึกษาปัจจัยที่มีผลต่อราคา และ สร้างเครื่องมือที่ใช้ทำนายราคาBitcoinเป็นรายสัปดาห์โดยใช้ข้อมูลในการวิเคราะห์สร้างเป็นMachine learning model เพื่อช่วยตัดสินใจให้กับผู้ที่สนใจในการลงทุนBitcoin และ ลดความเสี่ยงในการลงทุน

## 1. ทำความเข้าใจธุรกิจและเป้าหมาย
ลักษณะธุรกิจเป็น ธุรกิจการเงินในด้านการลงทุนสินทรัพย์ดิจิทัล คือ Bitcoin ที่มีเสี่ยงและความผันผวนสูง โดยต้องการค้นหาปัจจัยที่สำคัญและทำนายbitcoinเป็นรายสัปดาห์ เพื่อช่วยตัดสินใจในการลงทุนและลดความเสี่ยงในการลงทุนให้กับผู้ที่สนใจการลงทุนBitcoin

#### ปัญหา 
1. การลงทุนใน Bitcoin ผิดพลาดและมีความเสี่ยงสูง
2. จะรู้ได้อย่างไรว่าปัจจัยที่สำคัญมีอะไรบ้าง
3. จะมีแนวทางในการทำนายBitcoinเป็นรายสัปดาห์อย่างไร
  
#### เป้าหมายทางธุรกิจ
1. ต้องการลดความเสี่ยงในการลงทุนในBitcoin
2. ต้องการรู้ว่าปัจจัยที่สำคัญที่มีผลต่อราคาBitcoinมีอะไรบ้าง
3. ต้องการสร้างแนวทางในการทำนายราคาBitcoinเป็นรายสัปดาห์
  
#### เป้าหมายทางวิทยาศาสตร์ข้อมูล
1. สร้างโมเดลหรืออัลกอลิทึมในการทำนายBitcoinเป็นรายสัปดาห์อย่างอัตโนมัติ
2. ค้นหาปัจจัยสำคัญที่ส่งผลต่อราคาBitcoin เพื่อทราบถึงปัจจัยที่สำคัญและใช้วิเคราะห์ความสัมพันธ์ระหว่างปัจจัยได้

## 2. ทำความเข้าใจข้อมูล
* ปัจจัยที่นำมาวิเคราะห์มีทั้งปัจจัยมาจากบทความหรือข้อเท็จจริงต่างๆและปัจจัยที่ผู้เขียนสงสัยอยากนำมาวิเคราะห์ร่วมด้วยแล้วคาดว่าจะมีผลต่อราคาBitcoin โดยข้อมูลที่นำมาวิเคราะห์นำมาจากหลายแหล่งตามแต่ละปัจจัย และ รูปแบบของข้อมูลเราเป็น Time series data ทุกปัจจัยมีความสัมพันธ์กับช่วงเวลาเป็นรายสัปดาห์ โดยเหตุผลที่วิเคราะห์เป็นรายสัปดาห์ เพราะ บางปัจจัยไม่ได้มีการเคลื่อนไหวเป็นรายวัน เช่น interest rate(อัตราดอกเบี้ย), bond yield(ผลตอบแทนของพันธบัตรรัฐบาล),	inflation rate (อัตราเงินเฟ้อ) เป็นต้น   
  
* เราจึงต้องทำให้ข้อมูลของแต่ละปัจจัยสัมพันธ์กันให้ได้ผ่านความถี่ที่เหมาะสม นั่นคือ รายสัปดาห์(Weekly) บวกกับ ราคาของBitcoinนั้นมีความผันผวนสูงการจะมองเป็นรายวันนั้นยากต่อการคาดเดา เพราะ ปัจจัยต่างๆนั้นมีความผันผวนสูงซึ่งมีผลต่อราคาในระยะสั้นได้ การทำนายรายวันจึงอาจจะไม่มีประสิทธิภาพได้ เราจึงเฉลี่ยราคาและปัจจัยต่างๆเป็นรายสัปดาห์ซึ่งจะมีความแน่นอนกว่าและลดความเสี่ยงได้ดีกว่าการลงทุนรายวัน ดังนั้นโปรเจคนี้จึงเหมาะกับผู้ที่จะลงทุนในระยะยาวเป็นรายสัปดาห์ หรือมากกว่านั้น ไม่เหมาะกับผู้ลงทุนระยะสั้นจากการเทรดเพื่อหากำไรระยะสั้น  
  
กลับมาที่ Datasetของเราประกอบด้วย ตัวแปรหรือปัจจัยต่างๆ มีรายละเอียดดังนี้   

**1) ราคาBitcoin (BTC_price)** **หน่วย: Dollar** [source](https://finance.yahoo.com/quote/BTC-USD/history?period1=1593157142&period2=1624693142&interval=1wk&filter=history&frequency=1wk&includeAdjustedClose=true) เป็นตัวแปรตาม(target variable)  
   
**2) มูลค่าสินทรัพย์ (Marketcap)** คือ มูลค่าทั้งหมดของBitcoinที่มีอยู่ในตลาดcrytocurrency	**หน่วย: Dollar** [source](https://www.blockchain.com/charts/market-cap)
  
**3) อัตราการขุดBitcoin (Hash_rate)** คือ อัตราการขุด(prove of work) bitcoinต่อวินาทีมีชื่อหน่วยเป็นTheta hashต่อวินาที **หน่วย: TH/s** [source](https://www.blockchain.com/charts/hash-rate)     

**4) การลดจำนวนเหรียญครึ่งหนึ่งจากการขุด (Halving)** คือ ระบบของ Bitcoin จะทำการลดจำนวนเหรียญที่เกิดใหม่ของ Bitcoin ต่อ 1 บล็อกต่อ 10 นาทีลงครึ่งหนึ่ง ในทุก ๆ 4 ปี **ข้อมูลเป็นDummy variable หากค่าเป็น0 แสดงถึง ปีนั้นมีการhalving หรือ 1 แสดงถึง ปีนั้นไม่มีการhalving** [source](https://bitcoinblockhalf.com/)     

**5) รางวัลจากการขุด (Block_reward)** คือ จำนวนBTC ทั้งหมดที่เกิดขึ้นจากการขุด ต่อ1บล็อกใน10นาที **หน่วย: BTC** [source](https://corporatefinanceinstitute.com/resources/knowledge/other/bitcoin-halving/#:~:text=Bitcoin%20halving%20is%20the%20process%20of%20halving%20the,tends%20to%20put%20upward%20pressure%20on%20its%20price.)     

**6) ค่าการทำธุรกรรมของBitcoin ในหน่วยBTC (Transaction_fee_btc)** คือ จำนวนBitcoinที่เกิดขึ้นจากการทำธุรกรรมในระบบblockchainของbitcoin โดยระบบของbitcoinจะถูกตั้งให้มีการเกิดขึ้นของbitcoin จากการทำธุรกรรมแล้วจะนำไปจ่ายเป็นรายได้ให้กับนักขุดbitcoinต่อไป **หน่วย: BTC** [source](https://www.blockchain.com/charts/transaction-fees)     

**7) ค่าการทำธุรกรรมของBitcoin ในหน่วยdollar (Transaction_fee_usd)** คือ มูลค่าUSD(dollar)ที่ถูกแปลงจากBitcoinที่เกิดขึ้นจากการทำธุรกรรมในระบบblockchainของBitcoin โดยระบบของbitcoinจะถูกตั้งให้มีการเกิดขึ้นของbitcoin จากการทำธุรกรรมแล้วจะนำไปจ่ายเป็นรายได้ให้กับนักขุดbitcoinต่อไป **หน่วย: Dollar** [source](https://www.blockchain.com/charts/transaction-fees-usd)	  

**8)  มูลค่าของตลาดDefi (Totalvaluelocked)** คือ มูลค่าของตลาดDecentralized finance โดยDefi หรือ Decentralized finance คือ ระบบการเงินใหม่ที่ไร้ตัวกลางอย่างธนาคาร เป็นเทคโนโลยีที่สามารถทำธุรกรรมทางการเงินได้ระหว่างลูกค้าด้วยกันได้อย่างอิสระ **หน่วย: Dollar** [source](https://defipulse.com/)

**14) จำนวนการทวิตเกี่ยวกับBitcoin (Tweets)** คือ จำนวนการทวิตและรีทวิตเฉลี่ยรายสัปดาห์ **หน่วย: ครั้ง** [source](https://bitinfocharts.com/comparison/bitcoin-tweets.html)

Exploratory Data Analysis หรือ EDA คือกระบวนการตรวจสอบ สำรวจข้อมูลเบื้องต้น ตรวจสอบสมมติฐานเบื้องต้น ตรวจสอบความผิดพลาดของชุดข้อมูล ก่อนนำไปวิเคราะห์ข้อมูลอย่างละเอียดหรือสร้างแบบจำลองทางสถิติ (Statistical Modeling)

### 2)  วิเคราะห์รายละเอียดทางสถิติของข้อมูล
เป็นการวิเคราะห์ข้อมูลเบื้องต้น เพื่อทราบถึง ค่าทางสถิติทั้ง count, mean, standard, deviation, minimun, maximum และ quartiles สิ่งที่ควรสังเกตหลักๆ คือ   

* meanหรือค่าเฉลี่ยแต่ละตัวแปรจะทำให้รู้ว่าเมื่อรวมทุกข้อมูลแล้วเฉลี่ยด้วยจำนวนข้อมูลมีค่าเท่ากับเท่าไหร่ สามารถรู้แนวโน้มของข้อมูลว่าส่วนใหญ่อยู่บริเวณไหนซึ่งจะใกล้เคียงกับค่าเฉลี่ยนั่นเอง    
  
* standard deviation หรือ ส่วนเบี่ยงเบนมาตรฐาน แสดงถึง การกระจายของข้อมูลจากค่าเฉลี่ย จะทำให้รู้ว่าข้อมูลกระจายมากน้อยขนาดไหน โดยถ้า stdยิ่งมีค่ามาก แสดงถึงการกระจายของข้อมูลที่มากนั่นเอง  
  
#### **แสดงค่าทางสถิติเบื้องต้น**

```python
#แสดงค่าทางสถิติเบื้องต้น
stat = data.describe()
display(stat)
```

จะเห็นว่าปัจจัยในกลุ่มของBitcoin ทั้งตัวแปร BTC_price, Marketcap, Hash_rate,Transaction_fee_btc และ Transaction_fee_usd ข้อมูลมีกระจายออกค่าค่าเฉลี่ยไปมาก สังเกตได้จากค่าStandard deviationที่สูง ยกตัวอย่างMarketcap มีค่าเฉลี่ยอยู่ที่ 1.829055e+11 dollar หรือ 1.82 * 10^11 dollar (จะมีการแปลงหน่วยในขั้นตอนต่อไปให้เข้าใจตัวเลขได้ง่ายขึ้น) และ มีการกระจายออกจากค่าเฉลี่ยถึง2.376527e+11 หมายความว่า มีข้อมูลที่มากกว่าค่าเฉลี่ยจนมีMarketcapอยู่ช่วง 4.205582e+11 dollarจำนวนมาก ซึ่ง 4.205582e+11 เกิดจากการนำค่าเฉลี่ย รวมกับ standard deviation  
  
ซึ่งไม่แปลกใจที่ข้อมูลBitcoin มีการกระจายตัวที่สูง เพราะ ความผันผวนและความเสี่ยงที่สูงเป็นคุณสมบัติของตลาดนี้อยู่แล้ว ดังนั้น เมื่อมีปัจจัยหนึ่งเปลี่ยนแปลงมากก็อาจจะส่งผลต่อปัจจัยหนึ่งมากเช่นกัน  
  
ขณะที่เมื่อเรามาดูปัจจัยอื่นที่ไม่ได้เป็นปัจจัยในกลุ่มของBitcoin จะมีค่าStandard deviation อยู่ในช่วงต่ำถึงกลางๆ ยกตัวอย่าง ดัชนีของตลาดหุ้นS&P, NASDAQ, SSEC และ ราคาทองคำ(gold price) มีการกระจายตัวอยู่กลางๆ อาจจะเป็นเพราะ ตลาดมีมูลค่าขนาดที่ใหญ่กว่าBitcoin และ มีความผันผวนหรือความเสี่ยงที่ไม่สูงมาก  

#### **แสดงGraphทางสถิติของตัวแปรที่สนใจ**

## 3) วิเคราะห์กราฟของข้อมูลเบื้องต้น

**BTC_price** มีการเติบโตเพิ่มขึ้นจากปี2016จนสูงสุดถึง 20000dollarในปี2017 หลังจากนั้นปรับตัวลดลงมาอยู่ช่วงต่ำกว่า10000dollar จนท้ายปี2020มีการเติบโตเพิ่มขึ้นอย่างมากถึง 60000dollar แล้วปรับตัวลดลงมาช่วง30000dollarในปี2021เดือน5

## 3. เตรียมพร้อมข้อมูล
การเตรียมข้อมูล โดยปกติแล้วถือเป็นขั้นตอนที่ใช้เวลานานพอสมควร เพราะ ต้องทำความสะอาดข้อมูลอย่างละเอียด และจัดรูปแบบของข้อมูลให้พร้อมใช้งาน ทั้งหมดเพื่อลดการเกิดปัญหาตามมาในขณะทำการสร้างโมเดล ทั้งข้อมูลเกิดbiasที่มาก, varianceที่สูง ซึ่งจะทำให้ผลการทำนายเกิดความคลาดเคลื่อนและไม่แม่นยำได้ การเตรียมพร้อมข้อมูลจึงเป็นขั้นตอนที่สำคัญ โดยมีขั้นตอนให้เห็นดังต่อไปนี้

### Missing value & Outlier
ในขั้นตอนก่อนหน้านี้เราได้ตรวจสอบการขาดหายของข้อมูล(missing value) แล้วพบว่า ไม่มีข้อมูลที่ขาดหายไป จึงไม่ต้องจัดการmissing value  
  
ในส่วนของoutlier หรือ ข้อมูลที่สุดโต่งนั้น จะเห็นได้จากปัจจัยที่มีการกระจายสูง หรือ standard deviationที่สูง จริงอยู่ว่าoutlierจะทำให้ข้อมูลมีความแปรปรวน (variance)ที่สูงและมีผลต่อการทำนายของโมเดลได้ แต่ลักษณะของข้อมูลนี้เป็นTime seies เป็นข้อมูลที่เกิดขึ้นจริงมีเพียงค่าเดียวในช่วงเวลาต่างๆ ซึ่งเป็นเสมือนข้อมูลประชากร ไม่ใช่กลุ่มตัวอย่าง โดยส่วนใหญ่outlierที่เกิดขึ้นในtime series data ล้วนมีเหตุผลหรือปัจจัยที่ส่งผลให้เกิดค่านั้นขึ้นมาอยู่เสมอ ดังนั้นจึง ไม่ต้องจัดการกับoutlierที่เกิดขึ้นในTime Series data ในกรณีนี้

### Hash_rate และ block_reward มีผลต่อราคาBitcoinอย่างไร

```python
#create function plot_time series
def plot_timeseries(axes, x, y, color, xlabel, ylabel) :
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)

#create function to set Xticks every 6 months
import matplotlib.dates as mdates
def set_xtick_every6month(x=6, ax=ax) :
    for i in range(2) : 
        fmt_half_year = mdates.MonthLocator(interval=x)
        ax.xaxis.set_major_locator(fmt_half_year)
```

เริ่มจากสร้าง function ที่ใช้สำหรับplot time series มีคำสั่งภายในคือ สร้างlineplot, ตั้งชื่อแกนx ,ตั้งชื่อแกนy และ เปลี่ยนแปลงสีของชื่อแกน ค่าภายในแกน และเส้นกราฟได้ และ สร้างอีกfunctionที่จะตั่งค่าให้แกนxแสดงช่วงเวลาในทุก6เดือนด้วย การสร้างfunctionจะทำให้เราสะดวกและประหยัดเวลาในการใช้คำสั่งหลายคำสั่งและหลายๆครั้งได้ เพียงแค่เรียกใช้function และใส่inputตามที่ตั่งค่าได้ ก็สามารถแสดงผลลัพธ์ได้อย่างรวดเร็ว สะดวกขึ้น 

```python
#plot Hash_rate and BTC_price together
fig, ax = plt.subplots()
plot_timeseries(ax, data.index, data['Hash_rate'], 'red', 'Date', 'Hash_rate (Million TH/s)')
# create twin axes that shares the x-axis
ax2 = ax.twinx()
plot_timeseries(ax2, data.index, data['BTC_price'], 'blue', 'Date', 'BTC_price (Dollar)')

#show Time of Block_reward
ax2.annotate('Block_reward 12.5 BTC', xy=(pd.Timestamp('2016-07-10'),0 ) \
            ,xytext=(pd.Timestamp('2016-03-15'),60000 ) ,          \
            arrowprops={'arrowstyle':'simple' ,'color':"orange"}      )
ax2.annotate('Block_reward 6.25 BTC', xy=(pd.Timestamp('2020-05-17'),0 ) \
            ,xytext=(pd.Timestamp('2020-01-26'),60000 ) ,          \
            arrowprops={'arrowstyle':'simple' ,'color':"orange"}      )

#set Xticks to show Years-Month every 6 months
set_xtick_every6month(x=6,ax=ax)
    
```

## 4.2 หาปัจจัยที่สำคัญต่อ ราคาBitcoin 
ปัจจัยต่างๆที่เลือกมามีอิทธิพลต่อราคาBitcoinไม่มากก็น้อย ในการtrainโมเดลหากมีปัจจัยที่มากเกินไปจะทำให้เกิดปัญหาoverfittingตามมาได้ซึ่งจะส่งผลให้ความแปรปรวนสูงขึ้นและการทำนายของโมเดลมีความคลาดเคลื่อนที่สูงขึ้นตามมา ดังนั้น เราจำเป็นจะต้องเลือกปัจจัยที่มีความสำคัญหรือมีอิทธิพลต่อราคาBitcoin ในการtrain model เพื่อลดปัญหาoverfitting และลดความคลาดเคลื่อนของโมเดลในการทำนาย

ก่อนจะค้นหาปัจจัยที่สำคัญเราจะต้องทำให้ข้อมูลมีคุณสมบัตินิ่ง(stationary)ก่อน เพราะ ข้อมูลอนุกรมเวลามักจะมีคุณลักษณะtrend หรือ seasonal ซึ่งจะทำให้ข้อมูลไม่stationary <<รายละเอียดจะอธิบายในขั้นตอนการสร้างโมเดลต่อไป>> จึงต้องทำdifferenceเพื่อให้ทราบผลกระทบที่แท้จริงระหว่างปัจจัย 

```python
# first difference
data_diff = data.diff().dropna().diff().dropna()
```

การจะทราบถึงปัจจัยที่สำคัญได้นั้นสามารถหาได้โดยการสร้างโมเดลขึ้นมาแล้วค้นหาความสำคัญระหว่างปัจจัยที่มีอิทธิพลต่อราคาBitcoin  เพราะ โมเดลที่ถูกtrain จะคำนวณหาความสัมพันธ์ของปัจจัยอื่นๆต่อตัวแปรตามนั่นก็คือ BTC_price แล้วความสัมพันธ์ของปัจจัยไหนมีความสำคัญต่อการอนุมัติสินเชื่อสามารถดูได้จากคำสั่ง feature_importances_ ซึ่งจะแสดงค่าความสำคัญของตัวแปรออกมาอยู่ในช่วง 0 ถึง 1 ยิ่งมีค่าสูง ยิ่งมีความสำคัญ

```python
# select feature data and target datab
X = data_diff.drop(columns=['BTC_price'])
Y = data_diff['BTC_price']

#train RandomForestClassifier model to find importance factor of BTC_price
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, max_depth = 10,random_state=1)
rf.fit(X,Y)

# examine feature importance
p_value = rf.feature_importances_

#Create a series with feature importances
featimp = pd.Series(p_value, index=data_diff.columns[1:]).sort_values()
print (featimp)
```

```python
#bar plot feature importance
plt.title('Feature Importances')
plt.barh(featimp.index, featimp, color='#ffd343')
plt.xlabel('Relative Importance')
plt.show()
```

**เมื่อเราเลือกตัวแปรที่มีค่าความสำคัญมากกว่า 0.0171 พบว่ามี 13 ตัวแปรที่มีความสำคัญ คือ Marketcap, Transaction_fee_btc, Hash_rate, Google_trend, Totalvaluelocked, Tweets, CPI, NASDAQ, Active_users, SSEC, S&P, Transaction_fee_btc และ Inflation_rate ตามลำดับ**

## **1) VAR Model หรือ Vector Autoregressive Model**  
เป็นTimeseries modelหนึ่งที่ใช้ทำนายข้อมูลอนุกรมเวลา(ในที่นี้คือ BTC_price) ขึ้นอยู่กับตัวแปรอิสระหลายตัวแปรย้อนหลังเป็นช่วงเวลา(Lagged Values of Endogenous Variables) และข้อมูลทุกตัวแปรจะต้องมีคุณสมบัตินิ่ง(Stationary)ก่อนนำมาสร้างโมเดล เพราะ ข้อมูลอนุกรมเวลามักจะมีคุณลักษณะtrend, seasonal, cycle ซึ่งข้อมูลจะไม่นิ่ง(stationary)แล้วจะทำให้เกิดความคลาดเคลื่อนที่สูงขึ้นได้ เราจึงต้องทำให้ข้อมูลstationaryก่อนนำมาสร้างโมเดล เพื่อให้การทำนายมีความแม่นยำและลดความคลาดเคลื่อน ต่างจากregression ที่ไม่สนใจคุณสมบัตินิ่งของข้อมูลและสร้างเป็นสมการเส้นตรงเลย   

### ทดสอบคุณสมบัตินิ่งของตัวแปร  
VAR model จำเป็นต้องใช้ข้อมูลที่มีความนิ่งแล้วก่อนนำมาสร้างโมเดล ดังนั้นเราต้องทดสอบคุณสมบัตินิ่งของตัวแปรก่อนด้วยวิธี Unit root test แล้วทำการแปลงข้อมูลให้อยู่ในรูปผลต่าง(difference) เพื่อให้ข้อมูลมีคุณสมบัตินิ่ง แล้วทดสอบunit rootอีกครั้ง หากยังไม่มีคุณสมบัตินิ่งอีกให้ทำการdifference อีกครั้ง จนกว่าจะมีคุณสมบัตินิ่ง

```python
# unit root test/Augmented Dickey-Fuller Test (ADF Test)
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

# create function to test and show result to be 'stationary or non-stationary'
def adf_test(ts, signif=0.05):
    dftest = adfuller(ts, autolag='AIC')
    adf = pd.Series(dftest[0:4], index=['Test Statistic','p-value','# Lags','# Observations'])

    for key,value in dftest[4].items():
        adf['Critical Value (%s)'%key] = value
    print (adf)
    
    p = adf['p-value']
    if p <= signif:
        print(f" Series is Stationary")
    else:
        print(f" Series is Non-Stationary")
```

```python
for c in df_train.columns: 
    print('Result of',c,'variable')
    adf_test(df_train[c])
    print('\n')
```

ผลการทดสอบนี้เกิดจากการทดสอบสมมติฐาน หากค่าtest statistic มีค่าน้อยกว่า critical valueที่ระดับนัยสำคัญ0.05 แล้วแสดงว่า ยอมรับสมมติฐานหลักที่ว่า ข้อมูลstationaryหรือมีคุณสมบัตินิ่ง ดังนั้น ข้อมูลส่วนใหญ่จึงไม่มีคุณสมบัตินิ่ง(non-sationary) เนื่องจาก มีคุณลักษณะtrendหรือseasonal ทำให้test statistic มีค่ามากกว่า critical value จึงไม่มีคุณสมบัติstationary แต่มีเพียง1ตัวแปรเท่านั้นที่stationary คือ Transaction_fee_usd, Active_users

### แปลงข้อมูลdifference และ standardscaler ให้กลับเป็นหน่วยเดิม
เมื่อได้ผลการทำนายมาเรียบร้อย เราจะสร้างฟังก์ชันสำหรับแปลงข้อมูลการทำนายBTC_Priceที่ถูกทำsecond difference และ แปลงstandardscalerให้กลับมาเป็นค่าตามหน่วยของตัวแปรBTC_Price คือ Dollar 

```python
# create function to invert_transformation
def invert_transformation(ds, df_forecast, second_diff=False):
    for col in ds.columns:
        # Undo the 2nd Differencing
        if second_diff:
            df_forecast[str(col)] = (ds[col].iloc[-1] - ds[col].iloc[-2]) + df_forecast[str(col)].cumsum()
        # Undo the 1st Differencing
        df_forecast[str(col)] = ds[col].iloc[-1] + df_forecast[str(col)].cumsum()
    return df_forecast

# invert_transformation
df_var_forecast = invert_transformation(df_train, df_var_forecast, second_diff=True)

#invert_standardscaler
df_var_forecast = scaler.inverse_transform(df_var_forecast)
df_var_forecast = pd.DataFrame(df_var_forecast, columns=df.columns, index=df.index[-n:])
display(df_var_forecast[['BTC_price']])
```

### สร้างฟังก์ชันสำหรับทำนายBTC_price รายสัปดาห์ ด้วย VAR Model
โดยฟังก์ชันนี้จะรวบคำสั่งทุกอย่างตั่งแต่ เลือกตัวแปรที่สำคัญ differenceข้อมูล, สร้างVAR model,fit model, predict value จนแปลงข้อมูลกลับให้อยู่ในหน่วยเดิม จะช่วยให้สะดวกเมื่อต้องการนำฟังก์ชันนี้ไปใช้งาน เพียงแค่ใส่input เป็นข้อมูล และ จำนวนสัปดาห์ที่ต้องการทำนาย โดยจำนวนสัปดาห์ควรไม่เกิน4สัปดาห์ เพราะ ยิ่งทำนายอนาคตไกลเท่าไรโมเดลจะทำนายคลาดเคลือนมากขึ้น เนื่องจาก เป็นการทำนายจากข้อมูลที่มีในปัจจุบันเท่านั้น หากมีข้อมูลอัพเดตเพิ่มขึ้นก็จะทำนายได้แม่นยำมากขึ้น

```python
def VAR_forecast_BTC_price(df_train, n_week):
    
    # preprocessing standardscaler
    scaler = StandardScaler()
    transforms = scaler.fit_transform(df_train)
    df_train = pd.DataFrame(transforms,columns=df_train.columns,index=df_train.index)
    
    # second difference data
    data_diff = df_train.diff().dropna()
    data_second_diff = data_diff.diff().dropna()

    # create VAR model and fit model
    var_model = VAR(data_second_diff)
    result_var_model = var_model.fit()

    # forecast BTC_price next n weeks
    lag_order = result_var_model.k_ar
    var_forecast = result_var_model.forecast(data_second_diff.values[-lag_order:], n_week)
    
    # create n_week
    df_trains = df_train.append(pd.DataFrame( index=pd.date_range(start=df_train.index[-1]\
                                             , periods=n_week+1, freq='W') ))
    
    # convert to dataframe
    df_var_forecast = pd.DataFrame(var_forecast, index=df_trains.index[-n_week:], columns=df_trains.columns)
    
    # invert_transformation 
    df_var_forecast = invert_transformation(df_train, df_var_forecast, second_diff=True)
    
    # invert_standardscaler
    df_var_forecast = scaler.inverse_transform(df_var_forecast)
    df_var_forecast = pd.DataFrame(df_var_forecast, columns=df_trains.columns \
                                   ,index=df_trains.index[-n_week:])

    # show forecast BTC_price next 2 weeks
    return df_var_forecast[['BTC_price']]
```

โดยฟังก์ชันนี้จะรวบคำสั่งทุกอย่างตั่งแต่สร้างVECM model,fit model ,predict value จนแปลงข้อมูลกลับให้อยู่ในหน่วยเดิม จะช่วยให้สะดวกเมื่อต้องการนำฟังก์ชันนี้ไปใช้งาน เพียงแค่ใส่input เป็นข้อมูล และ จำนวนสัปดาห์ที่ต้องการทำนาย โดยจำนวนสัปดาห์ควรไม่เกิน4สัปดาห์ เพราะ ยิ่งทำนายอนาคตไกลเท่าไรโมเดลจะทำนายคลาดเคลือนมากขึ้น เนื่องจาก เป็นการทำนายจากข้อมูลที่มีในปัจจุบันเท่านั้น หากมีข้อมูลอัพเดตเพิ่มขึ้นก็จะทำนายได้แม่นยำมากขึ้น

```python
def VECM_forecast_BTC_price(df_train, n_week):
    
    # preprocessing standardscaler
    scaler = StandardScaler()
    transforms = scaler.fit_transform(df_train)
    df_train = pd.DataFrame(transforms,columns=df_train.columns,index=df_train.index)
    
    # create VAR model and fit model
    vecm_model = VECM(endog = df_train, k_ar_diff = 2)
    result_vecm_model = vecm_model.fit()

    #forecast BTC_price next n weeks
    vecm_forecast = result_vecm_model.predict(steps=n_week)
    
    #create n_week
    df_trains = df_train.append(pd.DataFrame( index=pd.date_range(start=df_train.index[-1]\
                                             , periods=n_week+1, freq='W') ))
    
    #convert to dataframe
    df_vecm_forecast= pd.DataFrame(vecm_forecast, index=df.index[-n_week:] \
                                   ,columns=df.columns)
    
     # invert_standardscaler
    df_vecm_forecast = scaler.inverse_transform(df_vecm_forecast)
    df_vecm_forecast = pd.DataFrame(df_vecm_forecast, columns=df_trains.columns\
                                   , index=df_trains.index[-n_week:])
    
    # show forecast BTC_price next 2 weeks
    return df_vecm_forecast[['BTC_price']]
```

### สร้างฟังก์ชันสำหรับทำนายBTC_price รายสัปดาห์ ด้วย Prophet Model
โดยฟังก์ชันนี้จะรวบคำสั่งทุกอย่างตั่งแต่สร้างProphet model, fit model,สร้าง datetime สำหรับทำนายข้อมูล, predict value จนแปลงข้อมูลกลับให้อยู่ในหน่วยเดิม จะช่วยให้สะดวกเมื่อต้องการนำฟังก์ชันนี้ไปใช้งาน เพียงแค่ใส่input เป็นข้อมูล และ จำนวนสัปดาห์ที่ต้องการทำนาย โดยจำนวนสัปดาห์ควรไม่เกิน4สัปดาห์ เพราะ ยิ่งทำนายอนาคตไกลเท่าไรโมเดลจะทำนายคลาดเคลือนมากขึ้น เนื่องจาก เป็นการทำนายจากข้อมูลที่มีในปัจจุบันเท่านั้น หากมีข้อมูลอัพเดตเพิ่มขึ้นก็จะทำนายได้แม่นยำมากขึ้น

```python
def Prophet_forecast_BTC_price(df_train, n_week):
    
    # preprocessing standardscaler
    scaler = StandardScaler()
    transforms = scaler.fit_transform(df_train)
    df_train = pd.DataFrame(transforms,columns=df_train.columns,index=df_train.index)
    
    # change columns BTC_price to y and Date to ds
    df2 = df_train.copy()
    df2.reset_index(inplace=True)
    df2.rename(columns={'Date':'ds','BTC_price':'y'}, inplace=True)
    
    # create Prophet model and fit model
    prophet_model = Prophet()
    prophet_model.fit(df2)
    
    # make future dataframe n weeks
    future = pd.date_range(start='2021-06-13', periods=n_week, freq='W-SUN')
    future = pd.DataFrame(future, columns=['ds'])
    
    # forecast BTC_price next n weeks
    prophet_forecast = prophet_model.predict(future)
    
    # change column name and index to original
    prophet_forecast.rename(columns={'ds':'Date','yhat':'BTC_price'}, inplace=True)
    prophet_forecast.set_index('Date', inplace=True)
    
    #invert_standardscaler
    prophet_forecast2 = df_train.copy().iloc[-n_week:]
    prophet_forecast2['BTC_price'] = prophet_forecast['BTC_price'].values
    prophet_forecast2 = scaler.inverse_transform(prophet_forecast2)[:,0]
    
    #create n_week
    df_trains = df_train.append( pd.DataFrame( index=pd.date_range(start=df_train.index[-1]\
                                             ,periods=n_week+1, freq='W') ))
    
    #convert to dataframe
    df_prophet_forecast = pd.DataFrame( prophet_forecast2, index=df_trains.index[-n_week:] \
                                      ,columns=['BTC_price'])
    
    # show forecast BTC_price next 2 weeks
    return df_prophet_forecast
```

## 6. ประเมินผลลัพธ์จากการทำนาย ราคาBitcoinรายสัปดาห์
หลังจากสร้างโมเดลในการทำนายราคาBitcoin ทั้ง 3 โมเดลแล้วขั้นตอนต่อไปคือ การประเมินผลลัพธ์จากการทำนาย เพื่อที่จะทราบถึงความแม่นยำและความคลาดเคลื่อนในการทำนายข้อมูลของแต่ละโมเดล แล้วนำค่าความคลาดเคลื่อนของแต่ละโมเดลมาเปรียบเทียบกัน เพื่อวิเคราะห์และตัดสินใจเลือกโมเดลที่ดีที่สุดในการทำนายราคาBitcoin โดยการวัดผลของโมเดลที่เราจะใช้มีรายละเอียดดังนี้

* **RMSE (Root Mean Squared Error)** เป็นตัวชี้วัดความคลาดเคลื่อนของโมเดลในการทำนาย เป็นหน่วยเดียวกันกับข้อมูลที่ทำนาย เช่น RMSE = 5000 หมายความว่า การทำนายของโมเดลคลาดเคลื่อนจากความเป็นจริงไป 5000 Dollar เป็นต้น โดยตัวชี้วัดนี้เกิดจาก ค่าเฉลี่ยของผลต่างระหว่างค่าทำนายกับค่าจริง  ยิ่งมีค่าน้อยยิ่งแสดงถึงการทำนายที่ดี
  
 
* **MAPE (Mean Absolute Percentage Error)** เป็นตัวชี้วัดความคลาดเคลื่อนของโมเดลในการทำนาย แต่จะมีหน่วยเป็นเปอร์เซนต์ความคลาดเคลื่อน (%) เพื่อให้สะดวกในกาเข้าใจว่าโมเดลเราทำนายผิดพลาดไปกี่เปอร์เซนต์ เช่น MAPE = 15% หมายความว่า การทำนายของโมเดลคลาดเคลื่อนจากความเป็นจริงไป 15% เป็นต้น โดยตัวชี้วัดนี้เกิดจาก ค่าเฉลี่ยของผลต่างระหว่างค่าทำนายกับค่าจริงแล้วคิดเป็นเปอร์เซนต์ ยิ่งมีค่าน้อยยิ่งแสดงถึงการทำนายที่ดี

### ประเมินความคลาดเคลื่อนของ Vector Autoregressive model (VAR Model)

```python
# evaluate var model
# RMSE
from sklearn.metrics import mean_squared_error as MSE
RMSE_var = MSE(df_test['BTC_price'], df_var_forecast['BTC_price'])**(1/2)

# MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

MAPE_var = mean_absolute_percentage_error(df_test['BTC_price'], df_var_forecast['BTC_price'])
```

ทั้งนี้ ผลการทำนายของVAR Model อยู่ในระดับสูง(หากดูจากค่าความคลาดเคลื่อน) การพัฒนาโมเดลต่อไปให้มีความคลาดเคลื่อนน้อยลงอาจจะpreprocessingด้วยวิธีอื่น หรือ เลือกตัวแปรที่มีความสำคัญอื่นๆมาtrainโมเดลได้ ยิ่งโมเดลได้เรียนรู้กับปัจจัยที่มีความสำคัญมากเท่าไหร่จะทำให้ประสิทธิภาพการทำนายของโมเดลดีขึ้นได้ 

## 7. การนำไปใช้จริงและสรุปผลตามเป้าหมาย  
  

### การนำModelไปใช้จริง (หากนำไปใช้จริง)
การจะมาถึงขั้นตอนนี้ โมเดลจะต้องผ่านการประเมินผลมาเรียบร้อยแล้วและได้โมเดลที่ดีที่สุดออกมาในที่นี้คือ VAR Model แต่ยังไม่สามารถนำไปใช้งานจริงได้ เพราะ ในการทำงานจริงเราจะต้องสร้างระบบอื่นๆที่เกี่ยวข้องกับการใช้งานModelด้วย ไม่ว่าจะเป็น แหล่งข้อมูลที่ต้องนำเข้ามาในโมเดล, หน่วยประมวลผล, หน่วยแสดงผล, ที่เก็บข้อมูล เป็นต้น โดยหน้าที่หลักในการสร้างและรวมระบบเข้าด้วยกันเป็นของ Data Engineer หรืออาจจะมี Software Developer มาช่วยทำงานร่วมด้วย   

หลังจากที่สร้างระบบการใช้งานของโมเดลเรียบร้อยแล้วยังต้องผ่านการtestระบบก่อนทุกระบบ เช่น ระบบประมวลผล ระบบแสดงผล เป็นต้น รวมถึงการทดสอบModelด้วยเมื่อโมเดลทำงานเองอย่างอัตโนมัติ(Dynamic) เมื่อtestระบบสำเร็จสามารถใช้งานได้จริงและไม่เกิดปัญหาอะไรอีกในช่วงtestระบบ จึงนำระบบไปใช้จริงกับลูกค้าได้เพื่อช่วยตัดสินใจในการลงทุนBitcoin และ หากนำไปใช้จริงแล้วบังเอิญเกิดปัญหา ผู้ที่ออกแบบระบบด้านนั้นๆก็ต้องมาทำการแก้ไขและติดตามอีกครั้งจนกว่าจะแก้ไขได้สำเร็จแล้วนำไปใช้จริงได้อีกครั้ง ดังนั้น การนำไปใช้จริงไม่ใช่แค่สร้างโมเดลเสร็จหรือสร้างระบบเสร็จแล้วจบโปรเจค แต่ต้องติดตามการนำไปใช้จริงด้วย แล้วคอยแก้ไขปัญหาที่เกิดขึ้นต่อไป

ในฝั่งของData Scientist ก็ต้องติดตามการใช้งานจริงของโมเดลด้วยเช่นกัน แล้วคอยแก้ไขปัญหาที่เกิดขึ้นกับโมเดล และที่สำคัญคือ ต้องมีการอัพเดตโมเดลอยู่เรื่อยๆ อาจจะอาทิตย์ละครั้ง เป็นต้น โดยนำเข้าข้อมูลใหม่ในการtrain model เพราะ ปัจจัยต่างๆที่มีอิทธิพลต่อราคาBitcoinอาจเปลี่ยนแปลงไปได้ตามช่วงเวลา และเพื่อทำให้โมเดลได้เรียนรู้กับข้อมูลที่ใหม่มากขึ้นจะช่วยทำให้โมเดลมีการทำนายได้มีประสิทธิภาพหรือถูกต้องมากขึ้นได้   
  


### สรุปผลตามเป้าหมาย

#### เป้าหมายทางธุรกิจ
1. ต้องการลดความเสี่ยงในการลงทุนในBitcoin
2. ต้องการรู้ว่าปัจจัยที่สำคัญที่มีผลต่อราคาBitcoinมีอะไรบ้าง
3. ต้องการสร้างแนวทางในการทำนายราคาBitcoinเป็นรายสัปดาห์

#### ผลลัพธ์ที่คาดหวังทางธุรกิจ
1. สามารถลดความเสี่ยงในการลงทุนในBitcoinได้
2. ปัจจัยที่สำคัญที่มีผลต่อราคาBitcoin ได้แก่ มูลค่าสินทรัพย์(Marketcap), ค่าการทำธุรกรรมของBitcoin ในหน่วยBTC, อัตราการขุดBitcoin, แนวโน้มการค้นหาBitcoinในGoogle, มูลค่าของตลาดDefi, จำนวนการทวิตเกี่ยวกับBitcoin, ดัชนีราคาผู้บริโภคสหรัฐ, ดัชนีตลาดหุ้นNASDAQ, ปริมาณผู้ทำธุรกรรม, ดัชนีตลาดหุ้นจีน, ดัชนีตลาดหุ้นS&P, ค่าการทำธุรกรรมของBitcoin ในหน่วยUSD และ อัตราดอกเบี้ยสหรัฐ ตามลำดับ
3. สามารถสร้างแนวทางในการทำนายราคาBitcoinรายสัปดาห์ได้ โดยใช้โมเดลในการทำนายราคาBitcoin 

#### เป้าหมายทางวิทยาศาสตร์ข้อมูล
1. สร้างโมเดลหรืออัลกอลิทึมในการทำนายBitcoinเป็นรายสัปดาห์อย่างอัตโนมัติ
2. ค้นหาปัจจัยสำคัญที่ส่งผลต่อราคา Bitcoin เพื่อทราบถึงปัจจัยที่สำคัญและใช้วิเคราะห์ความสัมพันธ์ระหว่างปัจจัยได้

#### ผลลัพธ์ทางวิทยาศาตร์ข้อมูล
1. สร้างโมเดลที่ดีที่สุด คือ Vector Autoregressive model (VAR Model) ที่ใช้ทำนายราคาBitcoinได้อย่างอัตโนมัติ โดยมีค่าความคลาดเคลื่อนอยู่ที่ 2.81% 
2. ค้นพบปัจจัยที่สำคัญที่ส่งผลต่อ ฺBTC_price ได้แก่  Marketcap, Transaction_fee_btc, Hash_rate, Google_trend, Totalvaluelocked, Tweets, CPI, NASDAQ, Active_users, SSEC, S&P, Transaction_fee_btc และ Inflation_rate ตามลำดับ
